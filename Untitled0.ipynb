```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
child_dataset = pd.read_csv('/Age_Height_Data - Age_Height_Data.csv')
```

```python

child_dataset
```

```python
X = child_dataset[['Age in Months']]
Y = child_dataset['height in cm']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,Y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
#Perfect match of y and f
y =[10, 20, 30]
f =[10, 20, 30]
r2 = r2_score(y, f)
print('r2 score for perfect model is', r2)
```

```python
### Assume y is the actual value and f is the predicted values
# y and f not matched, only one value matched
y =[10, 20, 30]
f =[20, 20, 20]
r2 = r2_score(y, f)
print('r2 score for a model which predicts mean value always is', r2)
```

```python
# y and f not at all matched
y =[10, 20, 30]
f =[20, 10, 20]
r2 = r2_score(y, f)
print('r2 score for a model which predicts mean value always is', r2)
```
